```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
from tqdm import tqdm

from sklearn.datasets import load_svmlight_file
from sklearn.linear_model import Perceptron
from sklearn.datasets import load_svmlight_file
```

```python
X_train, y_train = load_svmlight_file('../exercise_4_linear/train.txt')
X_test, y_test = load_svmlight_file('../exercise_4_linear/test.txt')
```

```python
X_train_dense = X_train.toarray()
X_test_dense = X_test.toarray()
```

```python
x_plot = []
y_plot = []
len_train = len(X_train_dense)

for i in tqdm(range(100, len_train+1, 100)):
    train_index = np.random.randint(low=0, high=len_train, size=i)
    X_train_dense[train_index]
    clf = Perceptron(max_iter=5)
    clf.fit(X_train_dense[train_index], y_train[train_index])
    y_plot.append(clf.score(X_test_dense, y_test))
    x_plot.append(i)
```

```python
plt.figure(figsize=(16, 8))
plt.plot(x_plot, y_plot, linewidth=3)
plt.title('Test accuracy vs train size')
plt.xlabel('Train size')
plt.ylabel('Test accuracy')
```

São necessárias entre 2500 e 5000 amostras de treinamento para alcançar um score de ~90% na base de teste
